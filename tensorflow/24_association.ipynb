```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
df = pd.read_csv("./data/association_rule_data.csv")
df.head()
```

```python
ds = df.values
```

```python
ds[:3]
```

```python
ds.shape[0]
```

```python
tr = []

for i in range(ds.shape[0]):
    tr.append([str(ds[i][j]) for j in range(ds.shape[1])])
```

```python
tr[1]
```

```python
tr[0]
```

```python
df_tr = df.transpose()
ds_tr = df_tr.values
ds_tr[1]
```

```python
df_tr.head(2)
```

```python
rules = apriori(tr, min_support = 0.003, min_confidence = 0.2,
               min_lift = 5, min_length = 2)
results = list(rules)
```

```python
pd.DataFrame([results[0]])
```

```python
df_rule = pd.DataFrame(results)
df_rule.head()
```

```python
colnames = pd.DataFrame(np.array(results[0], dtype=object)[2]).columns.values.tolist()
colnames
```

```python
df_rule_stat = pd.DataFrame(df_rule.iloc[:, 2].apply(lambda x: x[0])).explode("ordered_statistics")
df_rule_stat["cols"] = colnames * (df_rule_stat.index.values.max() + 1)
df_rule_stat = df_rule_stat.reset_index()
df_rule_stat.head()
```

```python
df_rule_stat.head(20)
```

```python
df_rule_stat_pivot = df_rule_stat.pivot(index="index", columns="cols", values="ordered_statistics").reset_index()
df_rule_stat_pivot.head()
```

```python
df_rule_stat_pivot["support"] = df_rule["support"]
df_rule_stat_pivot.head()
df_rule_stat_pivot = df_rule_stat_pivot.drop(columns="index", axis =1)
```

```python
df_rule_stat_pivot = df_rule_stat_pivot.iloc[:, [4,0,1,2,3]]
df_rule_stat_pivot.head()
```
