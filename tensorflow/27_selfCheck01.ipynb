```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/player_stat.csv", 
                 sep=",")
df.head(2)
```

```python
df = df.dropna(subset=["Year"])
```

```python
df = df.astype({"Year":int})
```

```python
df
```

```python
index = df["Index"]
year = df["Year"]
```

```python
df= df.fillna(0)
df["Index"] = index
df["Year"] = year
df
```

```python
df_a = df
```

```python
df = pd.read_csv("./data/player_info.csv", 
                 sep=",",
                dtype={"Year_Start":int,
                      "Year_End":int,
                      "Age_Start":int})
```

```python
df
```

```python
df_b = df.dropna()
```

```python
targetPlayer = df_b.loc[(df_b["Year_Start"]) >= 1980 &
                        (df_b["Year_End"] - df_b["Year_Start"] + 1 >= 3)]
targetPlayer
```

```python
df = df_a.merge(targetPlayer, left_on = "Player", right_on="Name")
df
```

```python
df = df.drop(columns="Player", axis=1)
df
```

```python
df["Age"] = df["Year"] - df["Year_Start"] + df["Age_Start"]
```

```python
df.head(2)
```

```python
df["Main_Pos"] = df["Pos"].apply(lambda x:x[0:x.find("-")])
```
