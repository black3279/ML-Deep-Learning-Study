```python
targetData = df_baseball[(df_baseball["yearID"]>=2011) & (df_baseball["yearID"]<=2016)]
targetData = targetData[targetData["playerID"].isin(list)]
targetData = targetData[["playerID", "teamID"]].drop_duplicates()
pivotTab = targetData.pivot_table(index="playerID",
                            columns=["teamID"],
                           aggfunc="size",
                            fill_value=0
                           )
pivotTab
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
result = apriori(pivotTab,min_support=0.0015, use_colnames=True)
```

```python
analysisResult = association_rules(result,metric="confidence",
                 min_threshold=0.5)
analysisResult
```
