```python
import pandas as pd
import numpy as np
```

```python
usage_2016_1 = pd.read_csv("./data/usage_2016_1.csv")
usage_2016_2 = pd.read_csv("./data/usage_2016_2.csv")
usage_2016_3 = pd.read_csv("./data/usage_2016_3.csv")
```

```python
usage_2016_1.head(2)
```

```python
usage_2016_1.shape
```

```python
usage_2016 = pd.concat([usage_2016_1, usage_2016_2, usage_2016_3])
usage_2016.shape
```

```python
location = pd.read_csv("./data/location.csv")
location.head(2)
```

```python
usage_2016 = pd.merge(usage_2016,
                    location[["id","gu"]],
                    how = "inner",
                    left_on="id_out",
                    right_on="id")
usage_2016.head(2)
```

```python
usage_2016_gu = usage_2016.groupby("gu")["distance"].sum().reset_index()
usage_2016_gu
```

```python
usage_2017 = pd.read_csv("./data/monthly_usage_2017.csv")
usage_2017.head(2)
```

```python
usage_2017_gu = usage_2017.groupby("gu")["dist"].sum().reset_index()
usage_2017_gu
```

```python
usage_gu_total = usage_2016_gu.merge(usage_2017_gu,
                                    how="inner",
                                    left_on = "gu",
                                    right_on = "gu")
usage_gu_total
```

```python
from scipy.stats import ttest_rel
```

```python
stat, p = ttest_rel(usage_gu_total["distance"],
                   usage_gu_total["dist"])
round(abs(stat), 4)
```
