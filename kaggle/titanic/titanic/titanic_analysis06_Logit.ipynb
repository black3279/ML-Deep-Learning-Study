```python
df = pd.read_csv('./train.csv')
df.head(10)
```

```python
list = []
def classification(x):
        if(x[0] == 0 & x[1] == 0) :
            return 1
        else :
            return 0

df["alone"] = df[["SibSp", "Parch"]].apply(lambda x: classification(x), axis=1)
df
```

```python
df_x = df[["PassengerId","Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
df_x
```

```python
df_x = df_x.dropna(how = 'any')
df_x = df_x.reset_index(drop=True)
```

```python
df_y = df_x[["Survived"]]
df_x = df_x[["Pclass","Age","SibSp", "Parch","Fare"]].join(pd.get_dummies(df_x[["Sex","Embarked"]]))
```

```python
df_x
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression(max_iter=1000).fit(df_x, df_y.values.ravel())
```

```python
df_test_x = pd.read_csv("./test.csv")
df_test_x = df_test_x[["PassengerId","Pclass","Age","SibSp","Parch","Fare", "Cabin","Sex","Embarked"]]
df_test_y = df_test_x[["PassengerId"]]
df_test_x
```

```python
df_test_x["alone"] = df_test_x[["SibSp", "Parch"]].apply(lambda x: classification(x), axis=1)
df_test_x
```

```python
df_test_x = df_test_x[["Pclass","Age","SibSp", "Parch","Fare"]].join(pd.get_dummies(df_test_x[["Sex","Embarked"]]))
```

```python
df_test_x = df_test_x.fillna(df_test_x.mean())
```

```python
result = lr_model.predict(df_test_x)
```
